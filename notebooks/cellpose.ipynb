# CellposeSAM
A notebook to try out the API in the newest version of Cellpose

```python
import os
from pathlib import Path
from cellpose.models import CellposeModel
from dotenv import load_dotenv
import tifffile
from skimage.measure import regionprops_table

load_dotenv("../.env")
DATA_ROOT = Path(os.environ.get("DATA_ROOT"))
```

```python
# Load images
image_path = (
    DATA_ROOT / "processed/labeled/Au_01-vol_01/labeled-01/Au_01-vol_01-z_0425.tif"
)
label_path = image_path.parent / f"labels/{image_path.stem}-cells.tif"
image = tifffile.imread(image_path)

# Calculate equivalent cell diameter to initialize cell size in model
labels = tifffile.imread(label_path)
regions = regionprops_table(labels, properties=["equivalent_diameter_area"])
diam_mean = regions["equivalent_diameter_area"].mean()
print(f"Average diameter {diam_mean}")
```

```python
cellpose_model = CellposeModel(gpu=True)
output = cellpose_model.eval(image, batch_size=128)
tifffile.imwrite(f"{image_path.name}-cellposeSAM-pretrained.tif", output[0])
```
