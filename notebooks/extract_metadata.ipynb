```python
from pathlib import Path
import json
import pandas as pd
from imageio.v3 import immeta
```

```python
# Extract metadata to json file
def extract_metadata(image_path: Path) -> None:
    """Extract image metadata and write to json file."""
    metadata = immeta(image_path)
    matadata_dir = image_path.parent / 'metadata'
    matadata_dir.mkdir(exist_ok=True)
    metadata_file = matadata_dir / f'{image_path.stem}-metadata.json'
    with open(metadata_file, 'w') as f:
        json.dump(metadata, f, indent=4)


for image_path in Path('Data').rglob('*.tif'):
    extract_metadata(image_path)
```
